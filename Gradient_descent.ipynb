```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#plot a function
def plot_function(func):
    x_values = np.linspace(-10,10,1000)
#linspace command -10 is the start point and 10 is the end point and 1000 is total no.of values
    y_values = [func(x) for x in x_values]
# this funcn will be computed for every value of x in x_values
#  list comprehension is use to make our code compact and concise
#     print("x values are:", x_values)
#     print("y values are:", y_values)
    plt.plot(x_values,y_values)

plot_function(lambda x: x**2)

# lambda functions are the anonymous funcns without the name and the definition of that funcn is written in a single line
```

```python
plot_function(lambda x : np.sin(x))
```

```python
plot_function(lambda x: x**2 + 4*x + 3)
```

```python
def plot_derivative(func):
    x_values = np.linspace(-5,5,1000)
    delta_x = 0.0001
    y_values = (func(x_values + delta_x) -func(x_values))/delta_x
    plt.plot(x_values,y_values)

plot_derivative(lambda x: x**2)
```

```python
plot_derivative(lambda x :np.sin(x))
```

```python
plot_derivative(lambda x :np.cos(x))
```

```python
# Given a function f and a starting point w(weight),
#try to find the minima or gradient descent
#(when we will call the function ,it will keep on oscillating )

def gradient_descent(func,w):
    list_of_weights = []
    weight = w
    delta = 0.0001
    for i in range(100):
        derivative = (func (weight+delta) - func(weight))/delta
        weight = weight - derivative
        list_of_weights.append(weight)
    return list_of_weights
```

```python
gradient_descent(lambda x:x**2+4*x+3,10)
```

```python
def gradient_descent(func,w):
    list_of_weights = []
    weight = w
    delta = 0.0001
    learning_rate = 0.1
    for i in range(1000):
        derivative = (func (weight+delta) - func(weight))/delta
        weight = weight - (derivative*learning_rate)
        list_of_weights.append(weight)
    return list_of_weights
```

```python
gradient_descent(lambda x:x**2+4*x+3,10)
```

```python
def f(x):
    return x**2
x = np.linspace(-5,5,1000)
dfdx = np.gradient(f(x),x)
# gradient function is used to minimize the error funcn
plt.plot(x,f(x),label = 'f(x)')
plt.plot(x,dfdx,label = 'df/dx')
plt.legend()
plt.show()
```
