```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/House.csv")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df["mainroad"] = df.groupby("mainroad")["price"].transform("mean")
```

```python
df["guestroom"] = df.groupby("guestroom")["price"].transform("mean")
```

```python
df["basement"] = df.groupby("basement")["price"].transform("mean")
```

```python
df["hotwaterheating"] = df.groupby("hotwaterheating")["price"].transform("mean")
```

```python
df["airconditioning"] = df.groupby("airconditioning")["price"].transform("mean")
```

```python
df["prefarea"] = df.groupby("prefarea")["price"].transform("mean")
```

```python
df["furnishingstatus"] = df.groupby("furnishingstatus")["price"].transform("mean")
```

```python
df
```

```python
#normalization(scaling)
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df1 = pd.DataFrame(scaler.fit_transform(df),columns = df.columns)
df1
#MinMax scaler is the library which is used for feature scaling
```

```python
#axis =0 means rows will be drop access =1 means column will be drop

y=df1['price']
x=df1.drop("price",axis=1)
y.shape,x.shape
#shape command is used to tell the shape of the dataframe i.e how many rows and how many columns are there
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state =1)
x_train.shape,y_train.shape,x_test.shape,y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
x_pred = model.predict(x_train)
y_pred = model.predict(x_test)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.score(x_test,y_test)
```

```python
y_test_predict = model.predict(x_test)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
plt.scatter(y_test_predict,y_test)
plt.legend()
plt.show()
```

```python
import seaborn as sns
imp = pd.DataFrame(list(zip(x_test.columns,np.abs(model.coef_))),columns= ['Feature','coef'])
# zip command used to pack the diff datatypes together
sns.barplot(x='Feature',y='coef',data=imp)
plt.xticks(rotation=90)
```
